```python
import pandas as pd

# 读取文件
df = pd.read_csv("Traffic_Collision_Data_from_2010_to_Present_20251125.csv")

# 解析日期列（注意 Occurred）
df["Date Occurred"] = pd.to_datetime(df["Date Occurred"], format="%m/%d/%Y", errors="coerce")

# 提取年份
df["Year"] = df["Date Occurred"].dt.year

# 筛选年份 >= 2021
filtered_df = df[df["Year"] >= 2021]

# 显示前几行
display(filtered_df.head())

# 保存筛选后的文件
output_name = "Traffic_Collision_Data_2021_and_after.csv"
filtered_df.to_csv(output_name, index=False)


new_df = pd.read_csv("Traffic_Collision_Data_2021_and_after.csv")
print(new_df["Year"])

# 打印行数
print("筛选后共有行数:", len(filtered_df))
print("已保存为:", output_name)
```
